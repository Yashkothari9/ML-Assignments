## Implement a self-organizing feature map on IRIS data for clustering purpose and find the  Silhouette Coefficient of the clusters (Don’t use any inbuilt function). 

```
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=["Sepal length", "Sepal Width", "Petal Length", "Petal Width","label"])
print(data)
```

```
X=data[["Sepal length", "Sepal Width", "Petal Length", "Petal Width"]].to_numpy()
# print(X)
Yi=data[["label"]].to_numpy()
Y=[]
for i in Yi:
  if i=='Iris-setosa':
    Y.append([1,0,0])
  elif i=='Iris-versicolor':  
    Y.append([0,1,0])
  else:
    Y.append([0,0,1])  
print(Y)
```

```
def distance(X1,X2):
  return math.sqrt(((X1[0]-X2[0])**2)+((X1[1]-X2[1])**2)+((X1[2]-X2[2])**2)+((X1[3]-X2[3])**2))

def alpha(i,lr):
  return math.exp(-i/lr)

# def  

def SOM(X,epochs,lr,No,sigma):
  weights=np.random.rand(len(X[0]),3)#we are initializing weights with weight matrix of input_features*output_classes
  weights=weights.transpose()
  for _ in range(epochs):
    cluster=[[] for _ in range(3)]
    for i in range(len(X)):
      j=random.randint(0,150)
      dist=[]
      for wt in weights:
        dist.append(distance(X[j],wt))
      idx=dist.index(min(dist))  
      cluster[idx].append(X[j])
      for i in range(len(weights[0])):
        weights[idx][i]=weights[idx][i]+No*alpha(_,lr)*(X[j][i]-weights[idx][i])#weight updation in som
  return cluster


cluster=SOM(X,100,0.5,0.7,0.3)  
print(len(cluster[0])," ",len(cluster[1])," ",len(cluster[2]))
```

# Now finding silhouette coefficient

```
def silhouette_exp(cluster):
  silhouette=0
  count_sil=0
  for i in range(len(cluster)):
    if len(cluster[i])>0:
       point=cluster[i][random.randint(0,len(cluster[i]))]
       intra_distance=0
       count_intra=0
       for j in range(len(cluster[i])):
         intra_distance+=distance(point,cluster[i][j])
         count_intra+=1
       inter_distance=0
       count_inter=0
       for k in range(len(cluster)):
         if i!=k:
           for l in range(len(cluster[k])):
             count_inter+=1
             inter_distance+=distance(point,cluster[k][l])  
       intra_avg=intra_distance/count_intra
       inter_avg=inter_distance/count_inter
       silhouette=(inter_avg-intra_avg)/max(inter_avg,intra_avg)
       count_sil+=1
 
  sil_avg=silhouette/count_sil
  return sil_avg

som_1=silhouette_exp(cluster)
print('Silhouette Coefficient for user defind fuction of som:',som_1)
```
