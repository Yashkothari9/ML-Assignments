```
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=["Sepal length", "Sepal Width", "Petal Length", "Petal Width","label"])
print(data)
```

```
X=data[["Sepal length", "Sepal Width", "Petal Length", "Petal Width"]].to_numpy()
print(X.dtype)
Yi=data[["label"]].to_numpy()
Y=[]
for i in Yi:
  if i=='Iris-setosa':
    Y.append(0)
  elif i=='Iris-versicolor':  
    Y.append(1)
  else:
    Y.append(2)  
print(Y)
```

```
def distance(X1,X2):
  return (((X1[0]-X2[0])**2)+((X1[1]-X2[1])**2)+((X1[2]-X2[2])**2)+((X1[3]-X2[3])**2))**0.5 

def u_initialize(X,Y):
  size=len(X)
  u=np.zeros((3,size))       # u[i][j] ,i->classes(0,1,2) and j->(size of X)
  for i in range(size):
    u[Y[i]][i]=1              #assigning the u[i][j]=1 when the sample datapoint and the training point has same value else it remains 0
  return u

def k_nearest(trainX,x,k):
  neigh=[]
  for i in range(len(trainX)):
    d=distance(x,trainX[i])
    neigh.append([d,i])
  return sorted(neigh)[:k]  



def predict(testX,trainX,k,u):
  y_pred=[]
  for x in testX:
    neigh=k_nearest(trainX,x,k) 
    # print(neigh)
    maxi=-1
    idx=np.random.randint(0,3)
    for i in range(3):
      a=0
      b=0
      for l in range(k):
        if 1:
          d=1/(neigh[l][0]**2)
          a+=d*u[i][neigh[l][1]]
          b+=d
      if b==0:     
        print(i,x)
      # print(a/b, a,b)      
      if b!=0 and (a/b)>maxi:
        maxi=a/b
        idx=i
    y_pred.append(idx)
  return y_pred     

trainX,testX,trainY,testY=train_test_split(X, Y, test_size=0.3)
u=u_initialize(trainX,trainY)
y_pred=predict(testX,trainX,2,u)
print(testY)
print(y_pred)
acc_inbuilt=metrics.accuracy_score(testY, y_pred)*100
print("accuracy",acc_inbuilt)
```
