## 1) Create a mini calculator to know the age of a person after putting the range (i.e. date of birth and current date)

```
def birth_age(age):
  today=date.today()
  age=today.year - age.year - ((today.month, today.day) <(age.month, age.day)) 
  return age
print("enter you birth date:")         
year=int(input("year: "))
month=int(input("month: "))
day=int(input("day: "))
age=birth_age((date(year,month,day)))
print(age,"years")
```
