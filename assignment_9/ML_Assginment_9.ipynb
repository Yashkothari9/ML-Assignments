## SVM

### (a) Load IRIS dataset. Create a linearly separable dataset with two features of petal length and petal width and two classes ‘versicolor’ and ‘virginica’.

```
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=["Sepal length", "Sepal Width", "Petal Length", "Petal Width","label"])
print(data)
```

```
X=data[["Petal Length", "Petal Width"]].to_numpy()
Xi=X[50:,:]
# print(X)
Yi=data[["label"]].to_numpy()
Y=[]
for i in Yi:
  if i=='Iris-versicolor':  
    Y.append(0)
  elif i=='Iris-virginica':
    Y.append(1)  
Y=np.array(Y)    
print(Y)
print(len(Y))
```

## Using an inbuilt svm function

```
from sklearn.svm import SVC # "Support Vector Classifier" 
trainX,testX,trainY,testY=train_test_split(Xi, Y, test_size=0.3,random_state=np.random.randint(10,50))
clf = SVC(kernel='linear') 
clf.fit(trainX, trainY) 
```

```
y_pred = clf.predict(testX)
```

```
print("Accuracy:",metrics.accuracy_score(testY, y_pred)*100)
```

```
from sklearn.model_selection import KFold
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score
def class_acc(y_test,y_pred,cls):
  tot=0
  corr_pred=0
  for i in range(len(y_test)):
    if y_test[i]==cls:
      tot+=1
      if y_test[i]==y_pred[i]:
        corr_pred+=1
  if tot>0:    
    avg=(corr_pred/tot)*100
  else:
    avg=100
  return avg    


def evaluate_model(Xi, y):
 clf =SVC(kernel='linear') 
 cv =KFold(n_splits=5,random_state=42,shuffle=True)
 score=[]
 vesicolor_acc=[]
 virg_acc=[]
 for train_index, test_index in cv.split(Xi):
  #  print("Train Index: ", train_index)
  #  print("Test Index: ", test_index,"\n")
   X_train, X_test, y_train, y_test = Xi[train_index], Xi[test_index], Y[train_index], Y[test_index]
   clf.fit(X_train, y_train)
   y_pred = clf.predict(X_test)
   vesicolor_acc.append(class_acc(y_test,y_pred,0)) #0 denotes that we are calculating accuracy for vesicolor
   virg_acc.append(class_acc(y_test,y_pred,1))      #1 denotes that we are calculating accuracy for virginica
   score.append(metrics.accuracy_score(y_test, y_pred)*100)
 return score,vesicolor_acc,virg_acc
    

score,vesicolor_acc,virg_acc=evaluate_model(Xi,Y)
print(score)
avg=sum(score)/len(score)
print("the average accuracy is-",avg)
avg_vesi=sum(vesicolor_acc)/len(vesicolor_acc)
avg_virg=sum(virg_acc)/len(virg_acc)
print("vesicolor accuracy is- ",avg_vesi)
print("verginica_accuracy is-",avg_virg)
```

# RBF

```
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=["Sepal length", "Sepal Width", "Petal Length", "Petal Width","label"])
print(data)
X=data[["Sepal length", "Sepal Width", "Petal Length", "Petal Width"]].to_numpy()
# print(X)
Yi=data[["label"]].to_numpy()
Y=[]
for i in Yi:
  if i=='Iris-setosa':
    Y.append([1,0,0])
  elif i=='Iris-versicolor':  
    Y.append([0,1,0])
  else:
    Y.append([0,0,1])  
print(Y)
```

### Implimenting K-means to get centroid points(5 CENTROID POINTS) for the hiddens layer

```
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=7).fit(X)
centroid_pos = k_means.cluster_centers_
print(centroid_pos)
```

### Since we have 5 centroid points we would have 5 neurons in the hidden layer

```
def distance(X1,X2):
  return math.sqrt(((X1[0]-X2[0])**2)+((X1[1]-X2[1])**2)+((X1[2]-X2[2])**2)+((X1[3]-X2[3])**2))


def beta_calc(centroid_pos):
  maxi = -sys.maxsize
  for i in range(0,len(centroid_pos)-1):
    for j in range(i,len(centroid_pos)):
      dis=distance(centroid_pos[i],centroid_pos[j])
      if maxi<dis:
        maxi=dis
  return maxi

def check(a):#check function will check in which of the three neurons will be having larger value and accordingly assigning 1 to that and 0 to others
  index=0
  max=0
  for i in range(len(a)):
   if a[i]>max:
    max=a[i]
    index=i
  arr=[]
  for i in range(len(a)):
    if index==i:
      arr.append(1)
    else:
      arr.append(0)
  return arr

def equal(a,b):
  for i in range(len(a)):
    if a[i]!=b[i]:
      return False
  return True

def Accuracy(X,Y,weights):#if on multiplying x*weights we get the same matrix as the resultant matrix the we increment the count and then in total we check whether the given is same or not
  count=0
  for x,y in zip(X,Y):
    predicted=np.matmul(x,weights)
    predictive=check(predicted)
    pred=np.asarray(predictive)
    if equal(pred,y):
      count+=1
  acc=(count/len(Y))*100
  return acc

def hidden_layer(x,centroid_point,beta):
 hidden_lyr=[]
 for i in range(len(centroid_point)):
   hidden_lyr.append(math.exp(-beta*distance(centroid_point[i],x)))
 return hidden_lyr

def hidden_to_final(X,Y,epochs,lr):
  weights=np.ones((len(X[0]),3))#we are initializing weights
  for _ in range(epochs):
    for x,y in zip(X,Y):
      # print(x.shape)
      # print(weights.shape)
      predict=np.matmul(x,weights)#predict will be containing the matrix multiplication of input(features+bias)(1x5) and weights(5x3)
      predictive=check(predict)#here we are sending the perdict to sigmoid which would send a matrix of sigmoid values and then sending it to the check function which would return a matrix of [1,0,0] or
                               #[0,1,0] or [0,0,1] ,this signifies that if a[0]=1 then sesota,a[1]=1 then vesicolor otherwise a[2]=1 then virginica
      pred=np.asarray(predictive)
      delta=y-pred
      weights+=[(lr * delta)*i for i in x]#here we are updating the weight value after each iteration  
  # print("The training accuracy",accuracy)
  return weights

def RBF(X,Y,centroid_point,beta,epochs):
  hidden_layr=[]
  for x in X:
    hidden_layr.append(hidden_layer(x,centroid_point,beta))
  weights=hidden_to_final(hidden_layr,Y,100,0.5)
  return weights

def RBF_test(X,Y,centroid_pos,beta,epochs):
  hidden_layr=[]
  for x in X:
    hidden_layr.append(hidden_layer(x,centroid_pos,beta))
  accuracy=Accuracy(hidden_layr,Y,weights)
  return accuracy

beta=math.sqrt(10)/beta_calc(centroid_pos) #hyperparameter
trainX,testX,trainY,testY=train_test_split(X, Y, test_size=0.3,random_state=np.random.randint(10,50))
weights=RBF(trainX,trainY,centroid_pos,beta,1000)  
acc=RBF_test(testX,testY,centroid_pos,beta,1000)
print(acc)
```
