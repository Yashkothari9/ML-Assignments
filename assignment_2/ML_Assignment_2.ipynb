# Part 1: Calculate the Mean, Median, Mode and Standard deviation

```
import pandas as pd
import numpy as np
import scipy as sy
```

## 1. Load the wine dataset from the given ‘wine.csv’ file into avariable (say data).



```
data=pd.read_csv('https://drive.google.com/uc?export=download&id=1EV_vQ1KSLiszmleCpGsFihloHNdimDRk')
```

##2. From the variable, extract the numerical information into a new variable, say scores.

```
scores=data[:].to_numpy()
print(scores)
print(scores.shape)
print(type(scores))
```

## 3. Calculate the row-wise,column-wise, and overall mean, median, mode, and standard deviation of the dataset. In this question write separate functions for each operation and pass the variable(scores) as the parameter to the function

```

def fun_mean(val):
    print("\n mean of score: ", np.mean(val))  
    print("\n mean of scores row wise: ",np.mean(val,axis=1))
    print("\n the size of row wise mean:",np.mean(val,axis=1).shape)
    print("\n mean of scores column wise :",np.mean(val,axis=0))
    print("\n the size of column wise mean: ",np.mean(scores,axis=0).shape)
fun_mean(scores)
```

```
def fun_median(val):
  print("\n median of score: ", np.median(val))  
  print("\n median of scores row wise: ",np.median(val,axis=1))
  print("\n the size of row wise median:",np.median(val,axis=1).shape)
  print("\n median of scores column wise :",np.median(val,axis=0))
  print("\n the size of column wise median: ",np.median(scores,axis=0).shape)

fun_median(scores)
```

```
from scipy import stats
def fun_mode(val):
  print("\n  mode of score: ", stats.mode(val,axis=None)[0])  
  print("\n mode of scores row wise: ",stats.mode(val,axis=1)[0])
  #print("\n the size of row wise mode:",stats.mode(val,axis=1).shape)
  print("\n mode of scores column wise :",stats.mode(val,axis=0)[0])
  #print("\n the size of column wise mode: ",stats.mode(scores,axis=0).shape)

fun_mode(scores)
```

```
def fun_standard_dev(val):
  print("\n standard deviation of score: ", np.std(val))  
  print("\n standard deviation of scores row wise: ",np.std(val,axis=1))
  print("\n the size of row wise standard deviation:",np.std(val,axis=1).shape)
  print("\n standard deviation of scores column wise :",np.std(val,axis=0))
  print("\n the size of column wise standard deviation: ",np.std(scores,axis=0).shape)
fun_standard_dev(scores)
```
